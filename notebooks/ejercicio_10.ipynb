#### Procesamiento de las rondas.

Definimos dos diccionarios para alamcenar informacion de los participantes, y  el otro para almacenar ganadores por ronda.

````python
# Inicializamos los diccionarios
participantes = {}
round_winner = {}

# Bucle de iteracion sobre rondas
for i,r in enumerate(rounds):
    
    # Procesa cada ronda
    round_points = fn.procesar_ronda(r, participantes)
    
    # Registra el ganador sobre cada ronda
    fn.registrar_ganador_ronda(participantes, round_winner, r["theme"], round_points)

    # Imprimimos la infomracion de la ronda
    print(f"Ronda {i+1} - {r['theme']}:")
    print(f"    Ganador: {list(round_winner[r['theme']].keys())[0]} ({list(round_winner[r['theme']].values())[0]})")
    
    # Ordenamos los paricipantes
    participantes = fn.ordenar_participantes(participantes)

    ## Imprimos la tabla
    # Encabezado de la tabla
    print("Tabla de posiciones global:")
    print(f"{'Cocinero':<15} {'Puntaje':^10} {'Rondas Ganadas':^14} {'Mejor Ronda':^14}")
    print("-" * 54)

    # Iteramos sobre los participantes, e imprimimos su contenido
    for p, v in participantes.items():
        print(f"{p:<15} {v["total_score"]:^10} {v["wining_rounds"]:^14} {v["max_points"]:^14}")

    print("-" * 54)   
    print("\n")
````

#### Promedios

Una vez que ya iteramos sobre todas las rondas, y tenemos el puntaje final de cada participante calculamos su promedio y lo agregamos como un campo dentro del diccionario que alamcena la informacion de los participantes.

````python
cantidad_de_rondas = len(rounds)

# Iteramos sobre todos los participantes
for p in participantes:
    
    participantes[p]["promedio"] = fn.definir_promedios(
        participantes[p]["total_score"], cantidad_de_rondas
    )
````

#### Imprimos la informacion

Imprimimos la informacion almacenda en el diccionario de particiapentes.
