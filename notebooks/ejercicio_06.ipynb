### Analisis de hastags.

Uitilizaremos la funcion `frecuencia_hashtag()`, la cual recibe por parametro un texto, y retorna un diccionario con la frecuencia de hastags. 

````python
# Texto con hashtgs
posts = [
"Arrancando el lunes con energía #Motivación #NuevaSemana",
"Terminé mi primer proyecto en Python #Python #Programación #OrgullosoDeMi",
"No puedo creer el final de la serie #SinSpoilers #SerieAdicta",
"Nuevo video en el canal sobre #InteligenciaArtificial y #Python",
"Entrenamiento de hoy completado #Fitness #Motivación #NoPainNoGain",
"Leyendo sobre #InteligenciaArtificial y el futuro del trabajo #Tecnología",
"Arranqué a estudiar #Programación por mi cuenta #Python #Autodidacta",
"Finde de lluvia, maratón de series #SerieAdicta #Relax",
"Workshop de #InteligenciaArtificial en la universidad #Tecnología #Programación"
]
````

````python
frecuencia_hashtags = fn.frecuencia_hashtag(posts)

# Obtenemos los hashtags unicos
hashtags_unicos = {hashtag: cantidad for hashtag, cantidad in frecuencia_hashtags.items() if cantidad == 1}

# Obtenemos los hashtags repetidos
hashtags_repetidos = {hashtag: cantidad for hashtag, cantidad in frecuencia_hashtags.items() if cantidad > 1}
hashtags_repetidos = dict(sorted(hashtags_repetidos.items(), key=lambda item: item[1], reverse=True))

for hashtag, cantidad in hashtags_repetidos.items():
    print(f"El hashtag {hashtag} se repitio una cantidad de {cantidad}")
print(f"La cantidad de hashtags es de: {len(hashtags_unicos) + len(hashtags_repetidos)}")
````
