## Sorteo de amigo invisble 

Utilizaremos la funcion `definir_amigo_invisibl()` la cual recibe un str con una lista de elementos separados por coma, y retorna los amigos invisbles separados por coma.

```python
# Solicitamos al usuario la lista
lista_de_amigos = input("Porfavor ingrese una lista de amigos separados por (,). Deben ser al menos 3")

# Utilizamos la funcion para definir el diccionario
lista_de_amigos_invisibles = fn.definir_amigos_invisibles(lista_de_amigos)

for amigo, amigo_invisible in lista_de_amigos_invisibles.items():
    print(f"A {amigo.title()} le toco {amigo_invisible.title()}")
```
