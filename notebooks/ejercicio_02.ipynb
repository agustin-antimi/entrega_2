##  Duración de una playlist

### Dada la siguiente playlist con canciones y sus duraciones en formato "m:ss":

```py
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
```
Calcule la duración total de la playlist en formato m:ss, y encuentre la
canción más larga y la más corta.

````python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
````

#### Duracion total

Utilizaremos la funcion `convert_to_seconds` en cada elemento de la lista, y luego los sumaremos, para finalmente utilizar la funcion `convert_to_time()`, que nos devolver el tiempo en formato [m:ss]

````python
# Obtenemos el total de segundos
total_segundos = sum(map(lambda x: fn.convert_to_seconds(x["duration"]), playlist))

print(f"El total de tiempo de la plyalist es de: {fn.convert_to_time_format(total_segundos)}")
````

#### Cancion mas larga

Utilizaremos una forma parecida a la anterior, siendo que utilizaremos una funcion `max()`.

````python
# Obtenemos la cancion mas larga
cancion_mas_larga = max(playlist, key=lambda x: fn.convert_to_seconds(x["duration"]))

print(f"La cancion mas larga es: {cancion_mas_larga["title"]}, con una duracion de: {cancion_mas_larga["duration"]}")
````

#### Cancion mas corta 
En vez de utilizar la funcion `max()`, usamos `min()`.

````python
# Obtenemos la cancion mas corta
cancion_mas_corta = min(playlist, key=lambda x: fn.convert_to_seconds(x["duration"]))

print(f"La cancion mas corta es: {cancion_mas_corta["title"]}, con una duracion de: {cancion_mas_corta["duration"]}")
````
