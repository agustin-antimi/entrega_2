### Cifrado cesar

Para el cifrado utilizaremos la funcion `cifrado_cesar()`, donde le pasaremos un str con un mensaje y un int con la cantidad de desplazamientos, la funcion retornara un str con el mensaje cifrado.

```python
mensaje = input("Escriba el mensaje a cifrar")

# Controlamos que el usuario introduzca un valore numerico
try:
    desplazamientos = int(input("Introduzca las unidades a desplazar"))
    print(f"La cantidad de unidades es de: ({desplazamientos})")
except:
    print("El valor ingresado no es numerico, se seteo la cantidad de desplazamientos en (2)")    

# Ciframos el mensaje
mensaje_cifrado = fn.cifrado_cesar(mensaje, desplazamientos)

print(f"MENSAJE ORIGINAL:\n{mensaje}")
print(f"\nMENSAJE CIFRADO:\n{mensaje_cifrado}")
```
