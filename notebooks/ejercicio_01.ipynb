#### Cantidad de palabras

> Esta funcion esta implementada para separar solo por espacios en blanco, por lo que

````python
print(f"La cantidad total de palabras es de: {fn.contar_palabras(text)}")
````

````python
promedio_por_linea = fn.prom_palabras(text)

print(f"La cantidad promedio de palabras por linea es de: {promedio_por_linea:.2f}")
````

Utilizremos la funcion `encima_prom()`

````python
lineas_encima_promedio = fn.encima_prom(text)

print("Las lineas por encima del promedio son: \n")
for linea in lineas_encima_promedio:
    print(f"{linea} ({fn.contar_palabras(linea)} palabras)")
````
